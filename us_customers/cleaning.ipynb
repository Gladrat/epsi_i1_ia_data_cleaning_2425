# Cleaning "Technique"

```python
import pandas as pd
```

```python
df = pd.read_csv(r'.\us_customers.csv', sep=";", encoding="utf-8")
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns="Not_Useful_Column")
df.columns
```

```python
def strip_col(col):
    return col.str.strip() if col.dtype == "object" else col

df.apply(strip_col)
```

```python
df.info()
```

```python
df.apply(lambda cell_value: cell_value.str.strip() if cell_value.dtype == "object" else cell_value)
```

```python
import re

str_cols = df.select_dtypes(include=['object', 'string']).columns
str_cols

# 

mask = df[str_cols].apply(
    lambda col: col.map(
        lambda x: not bool(re.match(r'^[A-Za-z0-9](.*[A-Za-z0-9])?$', str(x)))
    )
)
mask
```

```python
df_bad_lines = df[mask.any(axis=1)]
df_bad_lines
```

```python
df[~mask.any(axis=1)]
```

```python
df = df[~mask.any(axis=1)]#.copy()
df
```

```python
# supprimer tous les caractres non numriques
df["Phone_Number"] = df["Phone_Number"].str.replace(r'[\D]', '', regex=True)
df["Phone_Number"] = df["Phone_Number"].replace('', pd.NA)
df["Phone_Number"]
```

```python
df["Phone_Number"] = df["Phone_Number"].astype("string")
df["Phone_Number"] = df["Phone_Number"].fillna('0000000000')
df["Phone_Number"]
```

```python
# (123) 456-7890
    # 10 chiffres obligatoires
    # 0000000000

mask_valid = (
    (df["Phone_Number"] != "0000000000") &
    (df["Phone_Number"].str.fullmatch(r'\d{10}'))
)
mask_valid
```

```python
df.loc[mask_valid, "Phone_Number"] = df.loc[mask_valid, "Phone_Number"].map(
    lambda x: f"({x[:3]}) {x[3:6]}-{x[6:]}"
)
df
```

```python
df["Address"].str.split(',', expand=True)
```

```python
df[["Street address", "State", "Zip"]] = df["Address"].str.split(',', expand=True)
df
```
